```python
positions = {
    'Pos': ['DF', 'MF', 'FW', 'GK', 'FW,MF', 'MF,FW', 'NaN', 'DF,MF', 'MF,DF', 'DF,FW', 
            'CB', 'CM', 'FW,DF', 'CF', 'DM', 'RB', 'LB', 'RW', 'AM', 'LW', 'RM', 'SS', 'LM']
}
```

```python
position_mapping = {
    'GK': 'Goalkeeper',
    'DF': 'Defence', 'DF,MF': 'Defence', 'DF,FW': 'Defence', 'DF,MF': 'Defence', 'CB': 'Defence',
    'RB': 'Defence', 'LB': 'Defence', 'FW,DF': 'Defence', 
    'MF': 'Midfield', 'MF,FW': 'Midfield', 'MF,DF': 'Midfield', 'CM': 'Midfield',
    'DM': 'Midfield', 'RW': 'Midfield', 'AM': 'Midfield', 'LW': 'Midfield',
    'RM': 'Midfield', 'LM': 'Midfield',
    'FW': 'Forward', 'FW,MF': 'Forward', 'CF': 'Forward', 'SS': 'Forward',
    'NaN': 'Unknown'
}
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
position_mapping
```

```python
# Map positions to categories

df = cleaned_data.copy()
```

```python
# Map positions to categories using .loc
df.loc[:, 'Position'] = df['Pos'].apply(lambda x: position_mapping.get(x, 'Unknown'))

# One-hot encode the categories
encoder = OneHotEncoder(sparse=False)
encoded_categories = encoder.fit_transform(df[['Position']])

# Create a DataFrame with one-hot encoded columns
encoded_df = pd.DataFrame(encoded_categories, columns=encoder.get_feature_names_out(['Position']))

# Concatenate the original DataFrame with the one-hot encoded DataFrame
final_df = pd.concat([df, encoded_df], axis=1)

# Display the result
print(final_df)
```

```python
final_df = final_df.drop('Position_Unknown', axis=1)
```
